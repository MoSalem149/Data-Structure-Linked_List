```python
class Node:   # Data & Pointer
    def __init__(self,data):       #Constructor
        self.data=data
        self.next=None
        
    def get_data(self):           # Show Data
        return self.data
    
    def set_data(self,newData):  # Put New Data
        self.data=newData
        
    def get_next(self):     # Show next
        return self.next
    
    def set_next(self,newNext):   # Put New Next
        self.next=newNext
#test
temp=Node(93)
print(temp.get_data())
```
